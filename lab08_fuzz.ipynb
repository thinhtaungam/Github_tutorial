```python
from sklearn import linear_model
import pandas as pd
import numpy as np
from pandas import ExcelWriter
from pandas import ExcelFile
import math
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
import seaborn as sns

sns.set_style('whitegrid')
%matplotlib inline

ILPD = pd.read_excel('data.xlsx', sheet_name='Sheet1')
ILPD.head()
ILPD.shape
Y = ILPD['DIEMTS'].to_numpy()
Diem = ILPD.drop(['HO TEN', 'DIEMTS','KQ'], axis = 1)
X = Diem.to_numpy()


model = linear_model.LinearRegression()
model.fit(X,Y)

print(pd.DataFrame({"Diem: ":Diem.columns,"Coefficients":model.coef_}).sort_values(by='Coefficients'))
#Sai so
print(model.intercept_)
y_pred=model.predict(X)
print(y_pred)
trainScore = math.sqrt(mean_squared_error(y_pred,Y))
print(f"Train score is {trainScore}")
print(r2_score(Y,y_pred))
x_ax = range(886)
plt.scatter(x_ax,Y,s=5,color="blue",label="Diem Thuc")
plt.plot(x_ax,y_pred,lw=1.5,color="red",label="Diem Du Doan")
plt.legend()
plt.show()

plt.scatter(Diem['TOAN'].values,Y,color = 'blue')
plt.plot(Diem['TOAN'].values,y_pred,color='red')
plt.title('Hoi quy tuyen tinh')
plt.xlabel("Diem mon toan")
plt.ylabel("Diem tuyen sinh")
plt.show()
plt.scatter(Diem['ANH'].values,Y,color = 'blue')
plt.plot(Diem['ANH'].values,y_pred,color='red')
plt.title('Hoi quy tuyen tinh')
plt.xlabel("Diem mon ANH")
plt.ylabel("Diem tuyen sinh")
plt.show()

```
